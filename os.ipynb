```python
model = YOLO("yolo11n-cls.pt")
model.train(
    data="dataset",
    epochs=70,
    imgsz=224,
    batch=16
)
```

```python
from ultralytics import YOLO

model = YOLO("C:\\Files for any work\\python university\\year2-2\\CV\\newproject\\runs\\classify\\train\\weights\\best.pt")
class_names = ["Normal", "Osteoporosis"]

image_paths = [
    "dataset/test/normal/Normal 704.jpg",
    "dataset/test/osteoporosis/Osteoporosis 730.jpg",
    "dataset/test/normal/Normal 710.jpg"
    ]

plt.figure(figsize=(15,5))

for i, img_path in enumerate(image_paths):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = model(img)
    probs = results[0].probs.data.cpu().numpy()
    pred_idx = probs.argmax()
    label = class_names[pred_idx]
    confidence = probs[pred_idx] * 100
    color = "green" if label.lower()=="normal" else "red"
    plt.subplot(1, 3, i+1)
    plt.imshow(img_rgb)
    plt.title(f"{label} ({confidence:.1f}%)", color=color)
    text = f"N: {probs[0]*100:.1f}%\nO: {probs[1]*100:.1f}%"
    plt.text(10, 50, text,
             bbox=dict(facecolor='white', alpha=0.7))
    plt.axis("off")
    
plt.tight_layout()
plt.show()
```
